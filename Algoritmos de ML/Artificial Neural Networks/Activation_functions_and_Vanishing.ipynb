# ACTIVATION FUNCTIONS AND VANISHING GRADIENTS

## Contexto teórico


Hasta ahora he aprendido que las diferentes funciones de activación tienen diferentes tipos de curvas de salida características. Y, los valores de gradiente correspondientes o sus derivados también son diferentes. En ciertos casos, con valores de entrada más altos, la derivada comienza a llegar a cero, causando el problema del gradiente de desaparción. Asi pues,  compararemos las funciones de activación "Sigmoidea" y "ReLU" y exploramos sus derivados para gradientes "vanishing": 

__Activación sigmoide Function__
* La función sigmoide transforma los valores de entrada en un intervalo comprendido entre 0 y 1
* Produce una curva en forma de "S", por lo que es adecuada para salidas probabilísticas o tareas de clasificación binaria.
* Continuo, diferenciable y monotonamente creciente a través de su dominio (- , + ).
* Su derivada es simple de calcular: σ (z)=σ(z) (1 σ(z)). Esto es esencial para la retropropagación.
* Útil para las capas de salida donde se requieren probabilidades (por ejemplo, regresión logística).
* Permite límites de decisión no lineales, permitiendo a las redes neuronales aprender relaciones complejas.

  __Limitaciones__
    * Sufre el problema del gradiente de desaparición: para grandes entradas positivas o negativas, el gradiente se acerca a cero, ralentizando el aprendizaje en redes profundas.
    * Computacionalmente caro debido a la operación exponencial involucrada.


__ReLU (unidad lineal rectificada) Activación Function__

* ReLU devuelve el máximo de cero y el valor de entrada utilizando f(x)=máx (0,x)f(x)=máx(0,x).
* No lineal pero más simple que sigmoide; produce cero para las entradas negativas y el valor de entrada para las entradas positivas.
* La derivada es directa: f (x)=1f (x)=1 para x>0x>0, y f (x)=0f (x)=0 para x 0x 0.
* Mitiga el problema del gradiente de desaparición manteniendo un gradiente constante (1) para las entradas positivas, lo que permite una convergencia más rápida en redes profundas.
* Computacionalmente eficiente debido a su simplicidad.

  __Limitaciones__

    * Puede llevar a "neuronas muertas", donde las neuronas producen cero consistentemente debido a entradas negativas. Esto se puede mitigar usando variantes como Leaky ReLU.


#### Importar las liberías necesarias


```python
import numpy as np
import matplotlib.pyplot as plt
```

#### Definir la función de activación sigmoidea


```python

def sigmoid(z):
    return 1 / (1 + np.exp(-z))
```

#### Definir el gradiente de la función (derivada)


```python
def sigmoid_derivative(z):
    return sigmoid(z) * (1 - sigmoid(z))
```

#### Definir la función ReLU


```python
def relu(z):
    return np.maximum(0, z)
```

#### Definir el ReLU Gradiente


La función cogerá un valor z y dará aplicará la correspondiente derivada del tipo ReLU devolviendo el valor 1 para todos los z mayores que 0.

```python
def relu_derivative(z):
    return np.where(z > 0, 1, 0)
```

#### Definir el número de inputs

Serán 400 valores entre -10 y 10. El gradiente se calculará para cada uno de esos valores: 

```python

z = np.linspace(-10, 10, 400)
```


Ahora calculamos las derivadas para las activaciones Sigmoidea y de ReLU: 

```python
sigmoid_grad = sigmoid_derivative(z)
relu_grad = relu_derivative(z)
```


Por último, visualizamos las funciones y sus gradientes a través de los outputs: 

```python
# Visualizar las funciones de activación
plt.figure(figsize=(12, 6))

# Sigmoid y su gradiente
plt.subplot(1, 2, 1)
plt.plot(z, sigmoid(z), label='Sigmoid Activation', color='b')
plt.plot(z, sigmoid_grad, label="Sigmoid Derivative", color='r', linestyle='--')
plt.title('Sigmoid Activation & Gradient')
plt.xlabel('Input Value (z)')
plt.ylabel('Activation / Gradient')
plt.legend()

# ReLU y su gradiente
plt.subplot(1, 2, 2)
plt.plot(z, relu(z), label='ReLU Activation', color='g')
plt.plot(z, relu_grad, label="ReLU Derivative", color='r', linestyle='--')
plt.title('ReLU Activation & Gradient')
plt.xlabel('Input Value (z)')
plt.ylabel('Activation / Gradient')
plt.legend()

plt.tight_layout()
plt.show()
```

### Ejercicio práctico



Creción de la función tangente hiperbólica y su derivada: 

```python
def tanh(z):
    return np.tanh(z)

def tanh_derivative(z):
    return 1 - np.tanh(z) ** 2
```

### Ejercicio práctico 2



Visualizar y comparar el output de la función tanH con la función ReLU para 100 valores generados entre - 5 y 5: 

```python
# Generar los valores sintéticos
z = np.linspace(-5, 5, 100)

# OBtener los valores del gradiente
tanh_grad = tanh_derivative(z)
relu_grad = relu_derivative(z)


# Visualizar las funciones de activación
plt.figure(figsize=(12, 6))

# ReLu y su gradiente
plt.subplot(1, 2, 1)
plt.plot(z, relu(z), label='ReLU Activation', color='g')
plt.plot(z, relu_grad, label="ReLU Derivative", color='r', linestyle='--')
plt.title('ReLU Activation & Gradient')
plt.xlabel('Input Value (z)')
plt.ylabel('Activation / Gradient')
plt.legend()


# TanH y su gradiente
plt.subplot(1, 2, 2)
plt.plot(z, tanh(z), label='tanH Activation', color='g')
plt.plot(z, tanh_grad, label="tanH Derivative", color='r', linestyle='--')
plt.title('tanH Activation & Gradient')
plt.xlabel('Input Value (z)')
plt.ylabel('Activation / Gradient')
plt.legend()

plt.tight_layout()
plt.show()
```
